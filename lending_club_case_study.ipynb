### 1.2 Fixing Columns


1. Add column names: Add column names if missing
2. Delete blank columns
3. Check for misalligned columns
 


```python
# check for missing column name
df_loan.columns.isna()
```

```python
df_loan.columns
```

```python
# Check Nulls in entire data frame
df_loan.isnull().sum()
```

There are columns having all data as NA, delete those unnnesassary columns

```python
# drop columns having all NULL values
df_loan = df_loan.dropna(axis=1, how="all")
```

```python
df_loan.head()
```

```python
df_loan.shape
```

```python
df_loan.isnull().sum()
```

```python
# dropping columns with Null values > 10000
df_loan = df_loan.drop( ['desc','mths_since_last_delinq','mths_since_last_record','next_pymnt_d'],axis=1)
```

```python
df_loan.isnull().sum()
```

```python
df_loan.shape
```

```python
#To find misalligned columns,find the values of term column, since there are only unique values and no suspicious values there is no misallined rows
df_loan.term.value_counts()
```

```python
# find number of unique values in each columns
single_unique_col=[]
for i in df_loan.columns:
    print("No of unique values of {0} {1}".format(i,len(df_loan[i].unique()),))
    if len(df_loan[i].unique()) == 1:
        single_unique_col.append(i)
print(single_unique_col)
```

```python
# dropping columns with single unique value since these are of no importance to us
df_loan = df_loan.drop( single_unique_col,axis=1)
```

```python
df_loan.shape
```

```python
# since we want to analyse the driving factors for behind loan default, i.e. the variables which are strong indicators of default
# we can drop unnessasary columns i.e columns with string names like emp_title , url , title since these have more unique values
df_loan = df_loan.drop( ['emp_title','url','title','id','member_id'],axis=1)
```

```python
# analyse collections_12_mths_ex_med ,chargeoff_within_12_mths ,tax_liens  which has only 2 values
print(df_loan['tax_liens'].isnull().any())
print(df_loan['collections_12_mths_ex_med'].isnull().any())
print(df_loan['chargeoff_within_12_mths'].isnull().any())
```

```python
#dropping these columns as there are only 1 unique value excluding NULL
df_loan = df_loan.drop( ['tax_liens','collections_12_mths_ex_med','chargeoff_within_12_mths'],axis=1)
df_loan.shape
```

```python
#Loan status current means Applicant is in the process of paying the instalments, i.e. the tenure of the loan is not yet completed. These data is opf no importance to us. Therefore remove this data

df_loan = df_loan[df_loan.loan_status != 'Current']
```

### Summary of  fixing columns:
1. There are no missing column names

2. There were 54 columns with all NA values. These are removed

3. There were 4 columns with Null values > 10000,these are removed. In total 58 columns were removed

4. Since term column has only 2 values, there are no misalligned columns

5. There are 6 columns with 1 unique value, these are of least importance for analysis, hence these 6 columns are dropped


### 2 Standardising Values
### 2.1 Data type conversion

```python
df_loan.dtypes
```

```python
#int_rate should be of float datatype
df_loan.int_rate=df_loan.int_rate.apply(lambda x: x[:-1])
df_loan.int_rate=df_loan.int_rate.astype('float64')
```

```python
#revol_util should be of float datatype
df_loan.dropna(subset=['revol_util'],inplace=True)
df_loan.revol_util=df_loan.revol_util.apply(lambda x: x[:-1])
df_loan.revol_util=df_loan.revol_util.astype('float64')
```

```python
# convert issue_d column which is of object dtype to datetime
df_loan['issue_d']= pd.to_datetime(df_loan['issue_d'], format='%b-%y')
```

```python
# convert issue_d earliest_cr_line which is of object dtype to datetime
df_loan['earliest_cr_line']= pd.to_datetime(df_loan['earliest_cr_line'], format='%b-%y')
```

```python
# convert issue_d last_pymnt_d which is of object dtype to datetime
df_loan['last_pymnt_d']= pd.to_datetime(df_loan['last_pymnt_d'], format='%b-%y')
```

```python
# convert last_credit_pull_d last_pymnt_d which is of object dtype to datetime
df_loan['last_credit_pull_d']= pd.to_datetime(df_loan['last_credit_pull_d'], format='%b-%y')
```

```python
df_loan.dtypes
```

### 2.1 Derived Metrics

```python
# derive year and month separate columns from issue_d column
df_loan['issue_year']= pd.DatetimeIndex(df_loan['issue_d']).year
df_loan['issue_month']= pd.DatetimeIndex(df_loan['issue_d']).month
```

```python
# binning the annual income for easy analysis
df_loan["anuual_income_buckets"]=pd.qcut(df_loan.annual_inc,q=5,labels=["VL","L","M","H","VH"])
```
